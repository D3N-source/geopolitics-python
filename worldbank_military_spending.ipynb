```python
import matplotlib.pyplot as plt
```

```python
url = "https://api.worldbank.org/v2/en/indicator/MS.MIL.XPND.GD.ZS?downloadformat=csv"
```

```python
import zipfile
import io
import requests

response = requests.get(url)
zip_file = zipfile.ZipFile(io.BytesIO(response.content))

zip_file.namelist()
```

```python
df = pd.read_csv(
    zip_file.open("API_MS.MIL.XPND.GD.ZS_DS2_en_csv_v2_50.csv"),
    skiprows=4
)

df.head()
```

```python
countries = ["Ukraine", "Russian Federation", "Romania", "Poland", "Germany"]

df_selected = df[df["Country Name"].isin(countries)]

df_selected[["Country Name", "2020", "2021", "2022"]]
```

df_selected.set_index("Country Name")[["2020", "2021", "2022"]].T.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Military Spending (% of GDP)")
plt.ylabel("% of GDP")
plt.xlabel("Year")
plt.show()

## Key Insights

- Ukraine shows a sharp increase in military spending as a percentage of GDP starting from 2022, reflecting the direct impact of the war.
- Russia maintains high military expenditure levels, but without a proportional spike compared to Ukraine, indicating structural defense spending.
- Eastern European NATO members (Romania, Poland) display a gradual increase, consistent with NATO commitments and heightened security concerns.
- Germany shows a more moderate trend, with changes influenced by policy shifts rather than immediate conflict exposure.

## Geopolitical Interpretation

The data highlights how armed conflict and perceived security threats directly influence national budget priorities.
Eastern European countries exposed to the Russian-Ukrainian conflict demonstrate a reallocation of resources toward defense,
confirming the role of NATO and regional security dynamics in shaping fiscal policy.
